```python
import os, os.path
import numpy as np
from time import time
import pandas as pd
import math as m
import setup_runs as sr
import importlib
import matplotlib.pyplot as plt
import itertools
```

```python
#get the strategy and parameters table
df_strat = pd.read_csv(sr.fp_csv_attribute_strategy)
df_params = pd.read_csv(sr.fp_csv_parameter_ranges)
#strategies to keep
keep_strats = [0, 1]
#reduce
df_params = df_params[df_params["strategy_id"].isin(keep_strats)]
 
#split by strategy
dfp_0 = df_params[df_params["strategy_id"] == 0].copy()
dfp_1 = df_params[df_params["strategy_id"] == 1].copy()

#set merge fields
fields_merge = ["sector", "parameter", "time_series_id"]
fields_data = ["normalize_group", "trajgroup_no_vary_q", "min_2050", "max_2050"] + [str(x) for x in df_params.columns if x.isnumeric()] 
fields_sort = fields_merge + ["strategy_id"]

#do the comparison
dict_comp = sr.compare_params(dfp_0, dfp_1, fields_merge, fields_data)

#sets of NDC levers to exclude
dict_excl_by_new_strat = {
    2: set({"trajgroup_9", "medida_forestacion_aumentada", "medida_manejo_aumentada", "pmr_strategy"}),
    3: set({"trajgroup_9", "medida_forestacion_aumentada", "medida_manejo_aumentada"}),
    4: set({"trajgroup_9", "medida_forestacion_aumentada"}),
    5: set({"trajgroup_9"})
}

#new parameters out and new attribute table
df_new_params = [df_params]
header = list(df_params.columns)
st_att = []
#index of robust strategy
k = 1
#get the string of the NDC+ decription
str_ndcp = str(df_strat[df_strat["strategy_id"] == 1]["strategy_name"].iloc[0])

#loop over
for st in dict_excl_by_new_strat.keys():
    
    p_excl = []
    #get parameters to exlcude
    for p in dict_excl_by_new_strat[st]:
        p_excl = p_excl + [x for x in list(df_params["parameter"].unique()) if (p in x)]
    p_excl = set(p_excl)
    
    #identify the parameters to copy over from ndc+
    params_ndcp = dict_comp["params_diff"] - p_excl
    
    #initialize
    df_new = dfp_0.copy()
    df_new = df_new[~df_new["parameter"].isin(params_ndcp)]
    df_app = dfp_1.copy()
    df_app = df_app[df_app["parameter"].isin(params_ndcp)]
    
    #concatenate and update the strategy id
    df_new = pd.concat([df_new, df_app[df_new.columns]], axis = 0)
    df_new["strategy_id"] = np.array([int(st) for x in range(len(df_new))])
    
    #update the broader parameter table
    if len(df_new_params) == 0:
        header = list(df_new.columns)
    df_new_params.append(df_new[header])

    
    ##  update the strategy attribute table
    
    #generate strategy name and description
    st_nm = "NDC+ Alternative " + str(k)
    ign_ps = list(dict_excl_by_new_strat[st])
    ign_ps.sort()
    st_ds = ("%s; "*len(ign_ps))%tuple(ign_ps)
    st_ds = str_ndcp + " without " + st_ds[0:-2] + "."
    row = [int(st), st_nm, st_nm, st_ds, 1]
    st_att.append(row)
    
    #iterate on robust index
    k += 1

#new parameter table
df_new_params = pd.concat(df_new_params, axis = 0)
df_new_params = df_new_params.sort_values(by = fields_sort).reset_index(drop = True)
df_new_params = df_new_params[df_params.columns]

#new strategy attribute table
st_att = pd.DataFrame(st_att, columns = list(df_strat.columns))
df_strat = pd.concat([df_strat[df_strat["strategy_id"].isin(keep_strats)], st_att], axis = 0).sort_values(by = ["strategy_id"]).reset_index(drop = True)

#write
df_new_params.to_csv(sr.fp_csv_parameter_ranges.replace(".csv", "_withrobust.csv"), index = None, encoding = "UTF-8")
df_strat.to_csv(sr.fp_csv_attribute_strategy, index = None, encoding = "UTF-8")

```
