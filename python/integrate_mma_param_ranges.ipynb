```python
import setup_runs as sr
import pandas as pd
import numpy as np
import os, os.path
import importlib
import shutil
import matplotlib.pyplot as plt
```

```python
dir_analytica = os.path.join(os.path.dirname(sr.dir_cur), "analytica")
df_params = pd.read_csv(sr.fp_csv_parameter_ranges)
df_pp = pd.read_csv(sr.fp_csv_attribute_pyparams)

dict_read = {}
df_new = []
#files
for sec in ["forestry", "agriculture"]:
    dict_read.update({sec: {}})
    sec_up = sec.upper()
    for t in ["MMA", "Best_Science"]:
        fp_in = os.path.join(dir_analytica, sec, "parameter_ranges-" + sec_up + "-" + t + ".csv")
        df_tmp = pd.read_csv(fp_in, delimiter = ";", encoding = "Latin-1")
        dict_read[sec].update({t: df_tmp.copy()})
        df_new.append(df_tmp)
df_new = pd.concat(df_new, axis = 0)
df_new = df_new[df_new["sector"].isin(["Analytica_forestry", "Analytica_agriculture", "Analytica_all"])].reset_index(drop = True)

#update parameter names
dict_param_repl = sr.build_dict(df_pp[["pyparams_by_sector", "parameter"]])
df_new["parameter"] = df_new["parameter"].replace(dict_param_repl)
df_new["parameter"] = [x.lower().replace(" ", "_").replace("Ã±", "gn").replace("_2019", "") for x in list(df_new["parameter"])]
df_new["parameter"] = df_new["parameter"].replace({"sustitucion_y_degradacion": "sustitucion_degradacion"})

df_params_new = df_params[(~df_params["parameter"].isin(df_new["parameter"]))]
#df_params_new = 
```
